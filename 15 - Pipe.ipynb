```python
import pandas as pd
```

```python
path = 'datasets/employee_list.parquet'
df = pd.read_parquet(path)
```

```python
df.head()
```

```python
def substitute_boolean(x_df):
    mapping = {False: 0, True: 1}
    x_df['has_parking_space'] = x_df['has_parking_space'].map(mapping)
    return x_df

def split_name(x_df):
    x_df['first_name'] = x_df.name.str.split(' ').str[0]
    x_df['last_name'] = x_df.name.str.split(' ').str[1]
    return x_df.drop(columns='name')

def change_dtypes(x_df):
    to_string = ['company', 'country', 'department', 'first_name', 'last_name']
    for col in to_string:
        x_df[col] = x_df[col].astype('string')
    return x_df

df = df.pipe(substitute_boolean).pipe(split_name).pipe(change_dtypes).head()
df
```
