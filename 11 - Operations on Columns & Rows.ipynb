```python
import numpy as np
import pandas as pd
```

```python
path = 'datasets/employee_list.parquet'
df = pd.read_parquet(path)
df = df.iloc[:10, :4]
```

```python
df
```

# Renaming colums

#### Pass a dictionary containing old and new columnn names to the rename functions

```python
mapper = {
    'employee_number': 'Employee Number',
    'name': 'Name',
    'company': 'Comapny',
    'country': 'Country',
}
df.rename(columns=mapper, inplace=True)
```

#### Pass a function to the rename function

```python
df.rename(columns=str.lower)
```

```python
fn = lambda x: x.lower().replace(' ', '_')
df.rename(columns=fn)
```

#### Set df.columns equal to the list of new column names

```python
df.columns = ['employee number', 'name', 'company', 'country']
```

#### Set df.columns equal to a modified df.columns

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

#### Use a list comprehension

```python
df.columns = [col.strip().lower().replace(' ', '_') for col in df.columns]
```

#### Use set_axis method

```python
new_col_names = ['employee number', 'name', 'company', 'country']
df.set_axis(new_col_names, axis=1, inplace=True)
```

#### ADD PREFIX TO ALL COLUMNS

```python
df.add_prefix('X_').head(n=2)
```

#### ADD SUFFIX TO ALL COLUMNS

```python
df.add_suffix('_Y').head(n=2)
```
