```python
import pandas as pd
import numpy as np

ds = '../../datasets/pandas-cookbook/movie.csv'

movies = pd.read_csv(ds, delimiter=',')
```

#### 1. Find the five lowest budged movies from the top 100 scoring movies

```python
# solution 1
(movies[['movie_title', 'imdb_score', 'budget']]
 .nlargest(100, 'imdb_score')
 .nsmallest(5, 'budget'))
```

```python
# solution 2
(
movies[['movie_title', 'imdb_score', 'budget']]
.sort_values('imdb_score', ascending=False, kind='mergesort').head(100)
.sort_values('budget').head(5)
)
```

#### 2. Find the highest rated movie of each year

```python
(
movies[['movie_title', 'title_year', 'imdb_score']]
.sort_values(['title_year', 'imdb_score'], ascending=False)
.drop_duplicates(subset='title_year')
)
```

#### 3. Find the 3 highest rated movies of each year


```python
# solution 1
(
movies[['movie_title', 'title_year', 'imdb_score']]
.groupby('title_year')
.apply(lambda x: x.nlargest(3, 'imdb_score'))
.reset_index(drop=True)
.sort_values(['title_year','imdb_score'], ascending=False)
)
```

```python
# solution 2
(
movies[['movie_title', 'title_year', 'imdb_score']]
.groupby('title_year', as_index=False)
.apply(lambda df: df.sort_values('imdb_score', ascending=False).head(3))
.droplevel(0)
.sort_values(['title_year','imdb_score'], ascending=[True, False])
.assign(title_year=movies.title_year.astype('Int64'))
)
```
