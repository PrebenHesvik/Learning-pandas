```python
import pandas as pd
import numpy as np

path = 'datasets/employee_list.parquet'
df = pd.read_parquet(path)
df.head()
```

#### Finding the 5 highest paid employees

```python
df.nlargest(5, 'salary')
```

```python
# Method 2
df.sort_values('salary', ascending=False).head(5)
```

#### Finding the 5 highest salaries among the 20 oldest employees

```python
# Method 1
df.nlargest(20, 'age').nsmallest(5, 'salary')
```

```python
# Method 2
(
df
.sort_values('age', ascending=False)
.head(20)
.sort_values('salary')
.head(5)
)
```

#### Finding the highest paid employee of each age

```python
df.sort_values(['age', 'salary'], ascending=False).drop_duplicates(subset='age')
```

#### Finding the three highest paid employees of each age group

```python
# Method 1
(
df.
groupby('age')
.apply(lambda x: x.nlargest(3, 'salary'))
.reset_index(drop=True)
.sort_values(['age', 'salary'], ascending=False)
)
```

```python
# Method 2
(
df
.groupby('age', as_index=False)
.apply(lambda df: df.sort_values('salary', ascending=False).head(3))
.droplevel(0)
.sort_values(['age', 'salary'], ascending=False)
)
```

# Some series sorting methods
#### Argsort
- This method returns the Series consisting of the indices that would sort the Series values.

```python
df_sample = df.sample(10).set_index('name', drop=False)
age = df_sample.age
age
```

```python
age.argsort()
```

```python
age[age.argsort()]
```

#### Argmax
- Return int position of the largest value in the Series.

```python
age
```

```python
age.argmax()
```

```python
age[age.argmax()]
```

#### Argmin

```python
age.argmin()
```

```python
age[age.argmin()]
```

#### Idxmax
- Returns indices labels, not integers

```python
age.idxmax()
```

```python
age[age.idxmax()]
```

### Rank

```python
age = df.set_index('name').age.sample(10, random_state=1000)
```

```python
age.sort_values()
```

```python
age.rank().sort_values()
```

Setting method equal to anything else than 'average' changes the rank given to the values that are equal

```python
age.rank(method='min').sort_values()
```

```python
age.rank(method='max').sort_values()
```
