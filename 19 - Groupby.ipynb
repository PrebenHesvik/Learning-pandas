```python
import pandas as pd
```

```python
path = 'datasets/employee_list.parquet'
df = pd.read_parquet(path)
df.head()
```

#### Find average salary for each company

```python
df.groupby('company')['salary'].mean()
```

#### Find average salary for each department within each company and return as a dataframe

```python
df.groupby(['company', 'department'])[['salary']].mean()
```

#### Find max age for each department within each company and return as a dataframe

```python
df.groupby(['company', 'department'])[['age']].max()
```

#### Find min age, max age, mean age for each department for each company and return as dataframe

```python
df.groupby(['company', 'department']).agg(
    youngest_age=pd.NamedAgg(column='age', aggfunc='min'),
    oldest_age=pd.NamedAgg(column='age', aggfunc='max'),
    mean_age=pd.NamedAgg(column='age', aggfunc='mean')
)
```
